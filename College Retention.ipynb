```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

import os
from IPython.display import Image
from IPython.core.display import display, HTML
```

```python
os.chdir(r"C:\Users\kah32\Documents\GitHub\predicting-highered-retention")

import college_data
college = college_data.college

#college = pd.read_csv('data/college.csv', encoding='latin_1')
#college = college.set_index('UNITID')
#college.to_csv('data/college.csv')
college.reset_index(inplace=True)
college.drop_duplicates(subset = "UNITID", inplace=True)
college = college.set_index("UNITID")
college.head()
```

```python
#college['category'].value_counts()
#college = college[college['category'] == 6]
```

```python
college_str = college[['name', 'address', 'city', 'state', 'zipcode', 'highest_level_offered', 'HBCU', 'locale', 'category', 'county', 'category_desc']].copy()
college_num = college[map(lambda x: x not in college_str.columns, list(college.columns))].copy()
for col in college_num.columns:
    college_num[col] = pd.to_numeric(college_num[col])
    #college_num[col].fillna(np.mean(college_num[col]), inplace=True)
    college_num[col].fillna(college_num[col].median(), inplace=True)

college = college_str.join(college_num, on='UNITID', how='left')
college_num.head()
```

```python
college_num['total_in_tuition_fees'] = college['in_tuition'] + college['in_fees']
college_num['total_out_tuition_fees'] = college['out_tuition'] + college['out_fees']

college.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_score

from collections import defaultdict


from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso, BayesianRidge
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
target = 'retention'

y = college_num[target]
x = college_num.drop(columns=[target], axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
    
names_reg = ["Linear Regression", "Ridge Regression", "Lasso Regression", "Random Forest Regressor", 
             "GradientBoost", "Bayesian Ridge"]

regressors = [
        LinearRegression(),
        Ridge(),
        Lasso(),
        RandomForestRegressor(),
        GradientBoostingRegressor(),
        BayesianRidge(),
    ]

#mse_list = []
rmse_list = []
mae_list = []
#r2_list = []

results = defaultdict(int)
    
    #runs our models in a loop and pushed them to a pandas dataframe
for n, r in zip(names_reg, regressors):
    model = r.fit(x_train, y_train)
    results[n] = model.score(x_test, y_test)
    results_df = pd.DataFrame.from_dict(results, orient='index', dtype='float', columns=['R Squared'])
    
    #preds = model.predict(x_test)
    #mse = np.sqrt(mean_squared_error(y_test,preds))
    #mse_list.append(mse)
    
    scores = cross_val_score(model, x_train, y_train, cv=3, scoring='neg_mean_squared_error')
    #mse = np.mean(-scores)
    #mse_list.append(mse)
    rmse = np.mean(np.sqrt(-scores))
    rmse_list.append(rmse)
    #r2_list.append(scores.mean())
    
    scores = cross_val_score(model, x_train, y_train, cv=3, scoring='neg_mean_absolute_error')
    mae = np.mean(scores)
    mae_list.append(-mae)
        

#results_df['MSE'] = mse_list
results_df['RMSE'] = rmse_list
results_df['MAE'] = mae_list
#results_df['R2CVS'] = r2_list
results_df.sort_values(by=['R Squared'], ascending=False)
```

```python
target = 'retention'

y = college_num[target]
x = college_num.drop(columns=[target], axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

model = GradientBoostingRegressor()
model.fit(x_train, y_train)
model.score(x_test, y_test)

preds = model.predict(x)
```

```python
df = x.copy()
df = df.join(college['name'], on='UNITID')
df = df.join(college['retention'], on='UNITID')
df['predicted_retention'] = preds
df['diff'] = df['retention'] - df['predicted_retention']
df[['name', 'retention', 'predicted_retention', 'diff']]#.to_csv('data/output.csv')
#df['diff'].mean()
```

```python
college = college_str.merge(df, on='UNITID', how='right')
#college.head()
college[['name_x', 'city', 'state', 'locale', 'longitude', 'latitude', 'retention', 'predicted_retention', 'diff']]#.to_csv('data/retention.csv')
```
